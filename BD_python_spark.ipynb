Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```python
import pandas as pd
```

```python
ratings = pd.read_csv('u.data.csv', header=None, delimiter='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
ratings.head()
```

```python
movies = pd.read_csv('u.item.csv', header=None, delimiter='|', encoding='latin-1' , names=['movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film_Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci_Fi', 'Thriller', 'War', 'Western'])
movies.head()
```

Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
ratings['user_id'].value_counts().head(1)
```

Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

```python
rating_405 = ratings[ratings['user_id'] == 405]
rating_405.head()
```

Добавьте к датафрейму из задания 3 столбцы:

По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет

столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей


```python
merged = pd.merge(left=rating_405, right=movies, how='inner', left_on='item_id', right_on='movie_id')
merged.head()
```

```python
ratings_count = ratings[['rating', 'item_id']].groupby('item_id').count().reset_index()
ratings_count.head()
```

```python
ratings_count.rename(columns = {'rating': 'rating_count'}, inplace=True)
ratings_count.head()
```

```python
ratings_sum = ratings[['rating', 'item_id']].groupby('item_id').sum().reset_index()
ratings_sum.head()
```

```python
ratings_sum.rename(columns = {'rating': 'rating_sum'}, inplace=True)
ratings_sum.head()
```

```python
two_ratings = ratings_count.merge(ratings_sum, how='inner')
two_ratings.head()
```

```python
ratings_done = merged.merge(two_ratings, how='inner', left_on='item_id', right_on='item_id')
ratings_done.head()
```

```python
ratings_done.drop('item_id', axis=1, inplace=True)
ratings_done.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

Проведу небольшую подготовку данных.

Удалю ненужные столбцы, разделю дату и оставлю только год выпуска фильма.

```python
df_for_ml = ratings_done.drop(['user_id', 'timestamp', 'movie_id', 'movie_title', 'video_release_date', 'IMDb_URL'], axis=1)
df_for_ml.head()
```

```python
df_for_ml[['day', 'month', 'year']] = df_for_ml['release_date'].str.split('-', 2, expand=True)
```

```python
df_for_ml.drop(['release_date', 'day', 'month'], axis=1, inplace=True)
```

```python
df_for_ml.head()
```

```python
df_for_ml = df_for_ml.astype({'year': int}, errors='raise')
df_for_ml.dtypes
```

```python
df_for_ml.info()
```

```python
X = df_for_ml.iloc[ :, 1:]
X.head()
```

```python
y= df_for_ml.iloc[ :, :1]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test) 
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```

```python
# ml Linear Regression with PySpark
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler
```

```python
joined_df.columns
```

```python
joined_df.show(5)
```

```python
pipeline = Pipeline(stages=
                    [
                     VectorAssembler(inputCols=['unknown', 'Action', 'Adventure', 'AdveAnimationnture', 
                                                'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
                                                'Film_Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci_Fi', 'Thriller', 'War', 'Western'], outputCol="features")
                    ])
```

```python
piplineTrained = pipeline.fit(joined_df)
```

```python
joined_df = piplineTrained.transform(joined_df)
joined_df.show()
```

```python
joined_df = joined_df.select(['features', 'rating'])
joined_df.show(5)
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
train, test = joined_df.randomSplit([0.8, 0.2], seed=42)
```

```python
lr = LinearRegression(featuresCol = 'features', labelCol='rating')
lr_model = lr.fit(train)
```

```python
trainingSummary = lr_model.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
```

```python
test_pred = lr_model.transform(test)
test_pred.show(5)
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
lr_evaluator = RegressionEvaluator(predictionCol="prediction", \
                 labelCol="rating")
print("R Squared (R2) on test data = %g" % lr_evaluator.evaluate(test_pred))
```
