```python
for col in df:
  print(f'Максимум в столбце {col}: {df[col].max()}')
  print(f'Минимум в столбце {col}: {df[col].min()}')
  print(f'Размах в столбце {col}: {df[col].max() - df[col].min()}')
  print(f'Размах в столбце {col}: {df[col].max() - df[col].min()}')
  print(f'Среднее арифметическое в столбце {col}: {df[col].mean()}')
  print(f'Мода в столбце {col}: {df[col].mode()[0]}')
  print(f'Медиана в столбце {col}: {df[col].median()}')
  print(f'Стандартное отклонение в столбце {col}: {df[col].std()}')
  print(f'Дисперсия в столбце {col}: {df[col].var()}')
  print(f'Первый квартиль в столбце {col}: {df[col].quantile(0.25)}')
  print(f'Третий квартиль в столбце {col}: {df[col].quantile(0.75)}')
  iqr = df[col].quantile(0.75) - df[col].quantile(0.25)
  print(f'Межквартильный размах в столбце {col}: {iqr}', end='\n\n')

  lq = df[col].quantile(0.25) - (1.5 * iqr)
  hq = df[col].quantile(0.75) + (1.5 * iqr)

  df = df[df[col].notna()]

  remove_outliers = df[~df[col].between(lq, hq, inclusive=True)]
  print(f'Выбросы в столбце {col} {remove_outliers[col]}')

  print('---------------------------------------------------------', end='\n\n')
```

Задание 3. Работа с пропусками

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df = pd.read_csv('horse_data.csv', header=None, usecols=[0, 1, 3, 4, 5, 6, 10, 22], names=['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome'], na_values='?')
df.isna().sum()
```

```python
df.describe()
```

```python
# Заполнение пропусков по наличии или отсутствии операции логично сделать в разрезе значений возраста и боли по моде

df['surgery?'] = df['surgery?'].fillna(df.groupby(['Age', 'pain'])['surgery?'].transform(lambda x: x.mode()[0]))
```

```python
# Сделал предположение, что температура может изменяться от пульса т.к. в описании поля есть понятия шок, а в зависимости от шока меняется пульс

df['rectal temperature'] = df['rectal temperature'].fillna(df.groupby(['pulse'])['rectal temperature'].transform('median'))
```

```python
# Сделал предположение, что пульс зависит от различных температур и боли

df['pulse'] = df['pulse'].fillna(df.groupby(['rectal temperature', 'temperature of extremities', 'pain'])['pulse'].transform('median'))
```

```python
# Сделал предположение что частота дыхания зависит от боли и пульса

df['respiratory rate'] = df['respiratory rate'].fillna(df.groupby(['pain', 'pulse'])['respiratory rate'].transform('median'))
```

```python
# Сделал предположение, что температура конечностей зависит от возраста и была ли операция или нет

df['temperature of extremities'] = df['temperature of extremities'].fillna(df.groupby(['Age', 'surgery?'])['temperature of extremities'].transform('median'))
```

```python
# Сделал предположение, что боль зависит от того была ли операция

df['pain'] = df['pain'].fillna(df.groupby(['surgery?'])['pain'].transform(lambda x: x.mode()[0]))
```

```python
# Сделал предположение, что исход зависит от возраста и была ли операция

df['outcome'] = df['outcome'].fillna(df.groupby(['surgery?', 'Age'])['outcome'].transform(lambda x: x.mode()[0]))
```
